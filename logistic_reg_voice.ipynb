```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('voice.csv')
data
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
y = data['label']
x = data.drop('label',axis=1)
x.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=99)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logModel = LogisticRegression()
```

```python
logModel.fit(x_train,y_train)
```

```python
predictions = logModel.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print('Confusion matrix is')
cm1 = confusion_matrix(y_test,predictions)
print(cm1)
```

```python
from sklearn import metrics
print(metrics.accuracy_score(y_test,predictions))
```

```python
print('Classification report is')
print(classification_report(y_test,predictions))
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(data.corr(),annot=True)
```

```python
# skew and kurt,centroid and median,Q25 and centroid,meanfreq and median are highly correlated pairs
# we are ignoring skew and kurt and checking the accuracy 
```

```python
x1 = data.drop(['label','skew','kurt'],axis=1)
x1.info()
```

```python
x_train1,x_test1,y_train1,y_test1 = train_test_split(x1,y,test_size=0.2,random_state=99)
logModel = LogisticRegression()
logModel.fit(x_train1,y_train1)
predictions = logModel.predict(x_test1)
```

```python
print('Confusion matrix is')
cm = confusion_matrix(y_test1,predictions)
print(cm)
```

```python
from sklearn import metrics
print(metrics.accuracy_score(y_test1,predictions))
```

```python
print('Classification report is')
print(classification_report(y_test1,predictions))
```

```python
# Initially, the accuracy was 91.95% . On ignoring the most correlated pair of variables, the accuracy was 91.79%
```
