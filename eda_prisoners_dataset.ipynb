```python
new_data = data.drop('YEAR',axis="columns")
new_data
```

```python
new_data['total_benefitted'] = new_data.sum(axis=1,skipna = True,numeric_only=True)
new_data
```

```python
row = [{'STATE/UT':'Total','No. of Inmates benefitted by Elementary Education':0,'No. of Inmates benefitted by Adult Education':0,'No. of Inmates benefitted by Higher Education':0,'No. of Inmates benefitted by Computer Course':0,'total_benefitted':0}]
final = new_data.append(row,ignore_index=True,sort=False)
final
```

```python
final.iloc[35] = final.sum(axis=0,skipna = True,numeric_only=True)
final
```

```python
final.loc[35,'STATE/UT'] = 'Total'
final
```

```python
xvar = final['STATE/UT']
yvar = final['total_benefitted']
plt.barh(xvar,yvar)
plt.figure(figsize=(1,1))
#plt.show
```
